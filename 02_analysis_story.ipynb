# 02 Analysis Story

This notebook presents the project narrative in the same order used by the dashboard:

1. stock performance
2. risk profile
3. financial fundamentals
4. integrated insight

```python
from pathlib import Path
import sys
import plotly.express as px

ROOT = Path.cwd().resolve().parents[0]
SRC_DIR = ROOT / 'src'
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

from sp500_insights.data_pipeline import build_outputs, TICKER_COLORS
```

```python
bundle = build_outputs()
daily_stock = bundle.daily_stock
fundamentals = bundle.fundamentals
company_year = bundle.company_year
monthly_returns = bundle.monthly_returns
```

## Stock Performance

```python
fig = px.bar(
    company_year,
    x='ticker',
    y='cumulative_return',
    color='year',
    barmode='group',
    title='Annual Cumulative Return by Firm'
)
fig.show()
```

## Risk Profile

```python
fig = px.line(
    daily_stock.dropna(subset=['rolling_20d_volatility']),
    x='date',
    y='rolling_20d_volatility',
    color='ticker',
    color_discrete_map=TICKER_COLORS,
    title='20-Day Rolling Annualized Volatility'
)
fig.show()
```

## Financial Fundamentals

```python
fig = px.bar(
    fundamentals[fundamentals['year'] == 2024],
    x='ticker',
    y='sales',
    color='ticker',
    color_discrete_map=TICKER_COLORS,
    title='2024 Sales by Firm'
)
fig.show()
```

## Integrated Insight

```python
fig = px.scatter(
    company_year,
    x='roa',
    y='cumulative_return',
    color='ticker',
    color_discrete_map=TICKER_COLORS,
    symbol='year',
    size='sales',
    title='Annual Return vs ROA'
)
fig.show()
```
